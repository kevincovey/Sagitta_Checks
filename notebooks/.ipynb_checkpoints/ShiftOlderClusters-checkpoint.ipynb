```python
#create a function to append a copy of all the rows in the array, shifted to the requested distances
def add_dists(Catalog, start, end, step):
    
    #count the number of sources in the catalog
    n_sources = len(Catalog)
    print(n_sources)
    
    #first calculate the distance moduli for all the stars in the catalog
    distMods = 5.*np.log10(1000./Catalog['parallax'])-5.
    
    #now save the absolute mags in each filter
    Mg = Catalog['g'] - distMods
    Mbp = Catalog['bp'] - distMods
    Mrp = Catalog['rp'] - distMods
    Mj = Catalog['j'] - distMods
    Mh = Catalog['h'] - distMods
    Mk = Catalog['k'] - distMods
    
    #now create an array of distances we want to shift the stars to
    dists = np.arange(start,end,step)
    
    #and create an array of ones with length of our catalog to multiply things by to make arrays
    makeArray = np.ones(n_sources)
    fakeIDs = np.arange(n_sources)
    toSave = Catalog
    
    for i in range(len(dists)):
        #print(a)
        thisparallax = 1000./dists[i]
        theseparallaxes = makeArray*thisparallax
        thisDistMod = 5.*np.log10(dists[i]/10.)
        theseDistMods = thisDistMod*makeArray
        fake_ids = np.longlong(fakeIDs+makeArray*(i+1)*10000)
        #print(fake_ids[0])
        NewSources = Table( [fake_ids, Catalog['ra'], Catalog['dec'], theseparallaxes, 
                             Catalog['eparallax'], Mg+theseDistMods, Mbp+theseDistMods, 
                             Mrp+theseDistMods, Mj+theseDistMods, Catalog['ej'], 
                             Mh+theseDistMods, Catalog['eh'], Mk+theseDistMods, Catalog['ek'], 
                             Catalog['fc'], Catalog['cluster'], Catalog['actual_gaia']], 
                             names = ('source_id', 'ra', 'dec', 'parallax', 'eparallax', 'g', 'bp', 'rp', 
                               'j', 'ej', 'h', 'eh', 'k', 'ek', 'fc', 'cluster', 'actual_gaia'))
        #print(len(NewSources), len(toSave))
        #print(NewSources['g'][100], NewSources['parallax'][100], toSave['g'][100], toSave['parallax'][100])
        toSave = vstack([toSave, NewSources])
        #print(len(toSave))

        
    return toSave   
```

```python
####import Meingast catalog
Coronae = fitsio.read("../TestCatalogs/Coronae+2MASS+GaiaDR2.fits", 
                        columns = ['GaiaDR2', 'RA_ICRS', 'DE_ICRS', 'Dist_raw_', 'fc', 'Cluster', 'parallax',
                                   'parallax_error', 'pmra', 'pmra_error', 'pmdec', 'pmdec_error', 
                                   'phot_g_mean_mag', 'phot_bp_mean_mag', 'phot_rp_mean_mag', 'a_g_val',
                                   'Jmag', 'e_Jmag', 'Hmag', 'e_Hmag', 'Kmag', 'e_Kmag'], ext=1)

####find the number of sources in the raw catalog
print(len(Coronae['GaiaDR2']))

####no need to trim the catalog to those sources with positive parallaxes; 100% do! 
#usefulParallaxes = np.nonzero(Coronae['parallax'])
#print(len(usefulParallaxes[0]))

####sagitta will complain if there are duplicate GaiaIDs -- let's find them. -- none!  
####(at least, none now -- had to re-query because cross-match resulted in a few multi-matches. 
#print(len(Coronae['GaiaDR2']), len(np.unique(Coronae['GaiaDR2'])))
#values, counts = np.unique(Coronae['GaiaDR2'], return_counts=True)
#print(values[counts == 2])

####convert the filtered array into the column names Sagitta wants 
####(but save the gaia ID in a different index to link results)
UsefulCoronae = Table( [Coronae['GaiaDR2'], Coronae['RA_ICRS'], Coronae['DE_ICRS'], 
                        Coronae['parallax'], Coronae['parallax_error'], Coronae['phot_g_mean_mag'], 
                        Coronae['phot_bp_mean_mag'], Coronae['phot_rp_mean_mag'], 
                        Coronae['Jmag'], Coronae['e_Jmag'], Coronae['Hmag'], Coronae['e_Hmag'], 
                        Coronae['Kmag'], Coronae['e_Kmag'], Coronae['fc'], Coronae['Cluster'], Coronae['GaiaDR2']], 
                       names = ('source_id', 'ra', 'dec', 'parallax', 'eparallax', 'g', 'bp', 'rp', 
                               'j', 'ej', 'h', 'eh', 'k', 'ek', 'fc', 'cluster', 'actual_gaia') )
#print(len(UsefulSco))

####now write out the simple array for Sagitta to assess
UsefulCoronae.write('../SagittaInputs/UsefulCoronae.fits', format = 'fits', overwrite=True)


#now inflate with lots of distance shifted copies of itself
longerCoronae = add_dists(UsefulCoronae, 10, 500, 3)
#print(len(longerSco))

#now write it out so Sagitta can assess it.
longerCoronae.write('../SagittaInputs/LongerCoronae.fits', format = 'fits', overwrite=True)
```

So now we just need to read in LongerCoronae-sagitta.fits and do our analysis on it (as well as the initial inputs)

```python
## read in outputs; note no uncertainties since no Gaia photometric uncertainties
LongerCoronae_outputs = fitsio.read("../SagittaOutputs/LongerCoronae-sagitta.fits", 
                          columns = ['source_id', 'ra', 'dec', 'parallax', 'eparallax', 'g', 'bp', 
                                     'rp', 'j', 'ej', 'h', 'eh', 'k', 'ek', 'fc', 'cluster', 'actual_gaia', 
                                     'l', 'b', 'av', 'pms', 'age'], ext=1)

UsefulCoronae_outputs = fitsio.read("../SagittaOutputs/UsefulCoronae-sagitta.fits", 
                                columns = ['source_id', 'ra', 'dec', 'parallax', 'eparallax', 'g', 'bp', 
                                     'rp', 'j', 'ej', 'h', 'eh', 'k', 'ek', 'fc', 'cluster', 'actual_gaia', 
                                     'l', 'b', 'av', 'pms', 'age'], ext=1)
```

### Start by investigating the properties of the cluster stars at their actual distances. 

```python
#### Now check the membership probabilities of the cluster members at their actual distances
n_prob_bins = 10

#find fraction of cluster members with pms prob > 80%
high_prob = UsefulCoronae_outputs[UsefulCoronae_outputs['pms'] > 0.8]
low_prob = UsefulCoronae_outputs[UsefulCoronae_outputs['pms'] < 0.8]

print('frac. of >80% pms status: ', len(high_prob)/(1.*len(UsefulCoronae_outputs)))

#find how many members of each cluster are flagged
values, counts = np.unique(UsefulCoronae_outputs['cluster'][UsefulCoronae_outputs['pms'] > 0.8], return_counts=True)
cluster_values, cluster_counts = np.unique(UsefulCoronae_outputs['cluster'], return_counts=True)
print('Cluster members flagged as PMS > 80%:')
for i in range(len(values)):
    print('   ', values[i], '  ', counts[i],'    ', cluster_counts[i], '    ', counts[i]/cluster_counts[i])

print('     ')
print('     ')
#find fraction of cluster members with pms prob > 80%
evenhigher_prob = UsefulCoronae_outputs[UsefulCoronae_outputs['pms'] > 0.95]
lower_prob = UsefulCoronae_outputs[UsefulCoronae_outputs['pms'] < 0.95]

print('frac. of >95% pms status: ', len(evenhigher_prob)/(1.*len(UsefulCoronae_outputs)))

#find how many members of each cluster are flagged
values, counts = np.unique(UsefulCoronae_outputs['cluster'][UsefulCoronae_outputs['pms'] > 0.95], return_counts=True)
cluster_values, cluster_counts = np.unique(UsefulCoronae_outputs['cluster'], return_counts=True)
print('Cluster members flagged as PMS > 95%:')
for i in range(len(values)):
    print('   ', values[i], '  ', counts[i],'    ', cluster_counts[i], '    ', counts[i]/cluster_counts[i])
    
    
n, bins, patches = plt.hist([UsefulCoronae_outputs['pms']], n_prob_bins, range = [0.,1.], alpha = 0.5)
plt.xlabel('PMS prob')
plt.ylabel('N$_{stars}$')
plt.xlim(0,1)
plt.savefig('../plots/CoronaeProbs.png')
```

### Interesting.  
### at 95% confidence, less than 1% of the cluster members are flagged as PMS candidates at their actual distances.
### at 80% confidence, though, 10% of the cluster members are flagged as PMS candidates.  There is also a clear correlation between the cluster's age and the fraction of candidates (so up to 50% of younger clusters are flagged).

### show where each population lives in the CMD

```python
#### Check if there are systematic differences in the color-magnitude positions of members and non-members
plt.scatter(UsefulCoronae_outputs['g']-UsefulCoronae_outputs['k'], UsefulCoronae_outputs['g'] - (5.*np.log10(1000./UsefulCoronae['parallax'])-5.), alpha = 0.1, color = 'black')
plt.scatter(UsefulCoronae_outputs['g'][UsefulCoronae_outputs['pms'] > 0.8]-UsefulCoronae_outputs['k'][UsefulCoronae_outputs['pms'] > 0.8], UsefulCoronae_outputs['g'][UsefulCoronae_outputs['pms'] > 0.8] - (5.*np.log10(1000./UsefulCoronae['parallax'][UsefulCoronae_outputs['pms'] > 0.8])-5.), alpha = 0.2, color = 'blue')
plt.scatter(UsefulCoronae_outputs['g'][UsefulCoronae_outputs['pms'] > 0.95]-UsefulCoronae_outputs['k'][UsefulCoronae_outputs['pms'] > 0.95], UsefulCoronae_outputs['g'][UsefulCoronae_outputs['pms'] > 0.95] - (5.*np.log10(1000./UsefulCoronae['parallax'][UsefulCoronae_outputs['pms'] > 0.95])-5.), alpha = 0.2, color = 'red')
plt.xlabel('G - K')
plt.ylabel('M$_G$')
plt.ylim(17.5,-3)
plt.xlim(-1,7)
plt.savefig('../plots/CoronaeColorMag.png')
```

### check the age distribution of the candidates flagged in both cases.

### check for distance dependence in the inferred ages and pms probabilities

```python
#### Check if there are systematic differences in the inferred membership probabilities as a function of distance
plt.scatter(1000./LongerCoronae_outputs['parallax'], LongerCoronae_outputs['pms'], alpha = 0.005, color = 'red')
plt.xlabel('distance (pc)')
plt.ylabel('Pmem')
plt.ylim(0,1)
plt.xlim(0,500)
plt.savefig('../plots/CoronaeDistPMS.png')
```

### Hmm.  Seems that the model is reluctant to assign a high membership probability for sources within 100pc or so. 

```python
#### Check if there are systematic differences in inferred ages as a function of distance
labels = ['PMS > 80%','PMS > 95%']
colors = ['blue', 'red']
plt.scatter(1000./LongerCoronae_outputs['parallax'], LongerCoronae_outputs['age'], alpha = 0.005, label = 'PMS > 0%', color = 'grey')
plt.scatter(1000./LongerCoronae_outputs['parallax'][LongerCoronae_outputs['pms'] > 0.8], LongerCoronae_outputs['age'][LongerCoronae_outputs['pms'] > 0.8], alpha = 0.02, label = 'PMS > 80%', color = 'blue')
plt.scatter(1000./LongerCoronae_outputs['parallax'][LongerCoronae_outputs['pms'] > 0.95], LongerCoronae_outputs['age'][LongerCoronae_outputs['pms'] > 0.95], alpha = 0.5, label = 'PMS > 95%', color = 'red')

plt.legend()
plt.xlabel('distance (pc)')
plt.ylabel('log age (Myrs)')
plt.ylim(5,8)
plt.xlim(0,500)
plt.savefig('../plots/CoronaeDistAge.png')
```

### age structure as a function of distance as well.  Seems that the higher confidence sources are more likely to be young (which makes sense?)

```python
plt.scatter(LongerCoronae_outputs['pms'], LongerCoronae_outputs['age'], alpha = 0.001, label = 'd > 0 pc', color = 'grey')
plt.scatter(LongerCoronae_outputs['pms'][1000./LongerCoronae_outputs['parallax'] > 80], LongerCoronae_outputs['age'][1000./LongerCoronae_outputs['parallax'] > 80], alpha = 0.01, label = 'd > 80 pc', color = 'blue')
plt.scatter(LongerCoronae_outputs['pms'][1000./LongerCoronae_outputs['parallax'] > 170], LongerCoronae_outputs['age'][1000./LongerCoronae_outputs['parallax'] > 170], alpha = 0.02, label = 'd > 170 pc', color = 'red')
plt.xlabel('PMS probability')
plt.ylabel('log age (Myrs)')
plt.ylim(5,8)
plt.xlim(0,1)
plt.savefig('../plots/CoronaeAgeProb.png')
```

```python
#create a function to append a copy of all the rows in the array, shifted to the requested distances
def add_dists(Catalog, start, end, step):
    
    #count the number of sources in the catalog
    n_sources = len(Catalog)
    print(n_sources)
    
    #first calculate the distance moduli for all the stars in the catalog
    distMods = 5.*np.log10(1000./Catalog['parallax'])-5.
    
    #now save the absolute mags in each filter
    Mg = Catalog['g'] - distMods
    Mbp = Catalog['bp'] - distMods
    Mrp = Catalog['rp'] - distMods
    Mj = Catalog['j'] - distMods
    Mh = Catalog['h'] - distMods
    Mk = Catalog['k'] - distMods
    
    #now create an array of distances we want to shift the stars to
    dists = np.arange(start,end,step)
    
    #and create an array of ones with length of our catalog to multiply things by to make arrays
    makeArray = np.ones(n_sources)
    fakeIDs = np.arange(n_sources)
    toSave = Catalog
    
    for i in range(len(dists)):
        #print(i)
        thisparallax = 1000./dists[i]
        theseparallaxes = makeArray*thisparallax
        thisDistMod = 5.*np.log10(dists[i]/10.)
        theseDistMods = thisDistMod*makeArray
        fake_ids = np.longlong(fakeIDs+makeArray*(i+1)*10000000)
        #print(fake_ids[0])
        NewSources = Table( [fake_ids, Catalog['ra'], Catalog['dec'], theseparallaxes, 
                             Catalog['eparallax'], Mg+theseDistMods, Catalog['eg'], Mbp+theseDistMods, Catalog['ebp'],
                             Mrp+theseDistMods, Catalog['erp'], Mj+theseDistMods, Catalog['ej'], 
                             Mh+theseDistMods, Catalog['eh'], Mk+theseDistMods, Catalog['ek'], Catalog['source_id']], 
                             names = ('source_id', 'ra', 'dec', 'parallax', 'eparallax', 'g', 'eg', 'bp', 'ebp', 'rp', 'erp', 
                               'j', 'ej', 'h', 'eh', 'k', 'ek', 'actual_danceid'))
        #print(len(NewSources), len(toSave))
        #print(NewSources['g'][100], NewSources['parallax'][100], toSave['g'][100], toSave['parallax'][100])
        toSave = vstack([toSave, NewSources])
        #print(len(toSave))

        
    return toSave   
```

```python
####import DANCe catalog for IC4665 field (but restricted to objects with G < 18 and Pmem < 0.3, so confident field stars)
DANCe = fitsio.read("../TestCatalogs/DANCeIC4665_nonmembers.fits", 
                        columns = ['DANCe', 'RAJ2000', 'DEJ2000', 'Plx', 'e_Plx', 'pmRA', 'e_pmRA', 
                                   'pmDE', 'e_pmDE', 'Gmag', 'e_Gmag', 'BPmag', 'e_BPmag', 'RPmag', 'e_RPmag',
                                   'Jmag', 'e_Jmag', 'Hmag', 'e_Hmag', 'Ksmag', 'e_Ksmag'], ext=1)

####find the number of sources in the raw catalog
print(len(DANCe['DANCe']))

####no need to trim the catalog to those sources with positive parallaxes; 100% do! 
usefulParallaxes = np.nonzero(DANCe['Plx'])
print(len(usefulParallaxes[0]))

####sagitta will complain if there are duplicate GaiaIDs -- let's find them. -- none!  
####(at least, none now -- had to re-query because cross-match resulted in a few multi-matches. 
#print(len(Coronae['GaiaDR2']), len(np.unique(Coronae['GaiaDR2'])))
#values, counts = np.unique(Coronae['GaiaDR2'], return_counts=True)
#print(values[counts == 2])

####convert the filtered array into the column names Sagitta wants 
####(but save the gaia ID in a different index to link results)
UsefulDANCe = Table( [DANCe['DANCe'], DANCe['RAJ2000'], DANCe['DEJ2000'], 
                        DANCe['Plx'], DANCe['e_Plx'], DANCe['Gmag'], DANCe['e_Gmag'],
                        DANCe['BPmag'], DANCe['e_BPmag'], DANCe['RPmag'], DANCe['e_RPmag'],  
                        DANCe['Jmag'], DANCe['e_Jmag'], DANCe['Hmag'], DANCe['e_Hmag'], 
                        DANCe['Ksmag'], DANCe['e_Ksmag'], DANCe['DANCe']], 
                       names = ('source_id', 'ra', 'dec', 'parallax', 'eparallax', 'g', 'eg', 'bp', 'ebp', 'rp', 'erp', 
                               'j', 'ej', 'h', 'eh', 'k', 'ek', 'actual_danceid') )
#print(len(UsefulSco))

####now write out the simple array for Sagitta to assess
UsefulDANCe.write('../SagittaInputs/UsefulDANCe.fits', format = 'fits', overwrite=True)

# this catalog is large enough that I'm going to limit the grid of distance shifts

#now inflate with lots of distance shifted copies of itself
longerDANCe = add_dists(UsefulDANCe, 10, 500, 30)
#print(len(longerSco))

#now write it out so Sagitta can assess it.
longerDANCe.write('../SagittaInputs/LongerDANCe.fits', format = 'fits', overwrite=True)
```

    Generating Age uncertainties with 10 samples per star
        Scattering Age model inputs
        Running the age model on the varied data
        Aggregating the age uncertainties
    Saving output table to UsefulDANCe_results.fits
    Done

### Start by investigating the properties of the DANCe non-members at their actual distances. 

```python
#### Now check the membership probabilities of the cluster members at their actual distances
n_prob_bins = 10

#find fraction of cluster members with pms prob > 80%
high_prob = UsefulDANCe_results[UsefulDANCe_results['pms'] > 0.8]
low_prob = UsefulDANCe_results[UsefulDANCe_results['pms'] < 0.8]

print('frac. of >80% pms status: ', len(high_prob)/(1.*len(UsefulDANCe_results)))
print('num. of >80% pms status: ', len(high_prob))

#find fraction of cluster members with pms prob > 80%
evenhigher_prob = UsefulDANCe_results[UsefulDANCe_results['pms'] > 0.95]
lower_prob = UsefulDANCe_results[UsefulDANCe_results['pms'] < 0.95]

print('  ')
print('frac. of >95% pms status: ', len(evenhigher_prob)/(1.*len(UsefulDANCe_results)))
print('num. of >95% pms status: ', len(evenhigher_prob))

n, bins, patches = plt.hist([UsefulDANCe_results['pms']], n_prob_bins, range = [0.,1.], alpha = 0.5)
plt.xlabel('PMS prob')
plt.ylabel('N$_{stars}$')
plt.xlim(0,1)
plt.savefig('plots/DANCeProbs.png')
```

### Interesting.  
### at 95% confidence, very few field stars are spuriously flagged (<0.01%) as PMS candidates
### at 80% confidence, only 0.2% are flagged.  
### That is a very low rate, BUT -- there are a lot of stars in the galaxy, so absolute number matters too: 714 in this catalog.

### show where each population lives in the CMD

```python
#### Check if there are systematic differences in the color-magnitude positions of members and non-members
plt.scatter(UsefulDANCe_results['g']-UsefulDANCe_results['k'], UsefulDANCe_results['g'] - (5.*np.log10(1000./UsefulDANCe_results['parallax'])-5.), alpha = 0.1, color = 'black')
plt.scatter(UsefulDANCe_results['g'][UsefulDANCe_results['pms'] > 0.8]-UsefulDANCe_results['k'][UsefulDANCe_results['pms'] > 0.8], UsefulDANCe_results['g'][UsefulDANCe_results['pms'] > 0.8] - (5.*np.log10(1000./UsefulDANCe_results['parallax'][UsefulDANCe_results['pms'] > 0.8])-5.), alpha = 0.2, color = 'blue')
plt.scatter(UsefulDANCe_results['g'][UsefulDANCe_results['pms'] > 0.95]-UsefulDANCe_results['k'][UsefulDANCe_results['pms'] > 0.95], UsefulDANCe_results['g'][UsefulDANCe_results['pms'] > 0.95] - (5.*np.log10(1000./UsefulDANCe_results['parallax'][UsefulDANCe_results['pms'] > 0.95])-5.), alpha = 0.2, color = 'red')
plt.xlabel('G - K')
plt.ylabel('M$_G$')
plt.ylim(17.5,-3)
plt.xlim(-1,7)
plt.savefig('plots/DANCeColorMag.png')
```

```python
#### create an array of length UsefulSco to store ratio of actual spread to quoted uncertainties
UsefulDANCe_length = len(UsefulDANCe_results)
differences = np.zeros([UsefulDANCe_length, 17])
errors = np.zeros([UsefulDANCe_length, 17])
significances = np.zeros([UsefulDANCe_length, 17])

#### loop through the original UpperSco results
for i in range(UsefulDANCe_length):
        
    #find the ages of the distance shifted analogs by matching on the actual gaia id columns
    differences[i,:] = np.absolute((LongerDANCe_results['age'][LongerDANCe_results['actual_danceid'] == UsefulDANCe_results['actual_danceid'][i]]) - UsefulDANCe_results['age'][i])
    errors[i,:] = LongerDANCe_results['age_std'][LongerDANCe_results['actual_danceid'] == UsefulDANCe_results['actual_danceid'][i]]
    significances[i,:] = differences[i,:]/errors[i,:]
    
print(np.median(differences),np.median(errors),np.median(significances))
```
